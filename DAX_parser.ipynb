```python
# RU

import requests
from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET

sitemap_url = 'https://dax.co.il/product-sitemap.xml'
response = requests.get(sitemap_url)
sitemap_xml = response.content

root = ET.fromstring(sitemap_xml)
namespaces = {'ns': 'http://www.sitemaps.org/schemas/sitemap/0.9'}

products = []

for url in root.findall('ns:url/ns:loc', namespaces):
    product_url = url.text
    if product_url.startswith('https://dax.co.il/ru/tovar/'):
        product_page = requests.get(product_url)
        soup = BeautifulSoup(product_page.content, 'html.parser')

        modules = []
        for section in soup.find_all('section', class_='functions moreoptions'):
            h2 = section.find('h2')
            if h2 and h2.text.strip() == "Модули":
                gallery = section.find('div', id='sc_gallery')
                if gallery:
                    images = gallery.find_all('img')
                    image_links = [img['src'] for img in images]
                    image_texts = [img['alt'] for img in images]
                    for link, text in zip(image_links, image_texts):
                        if len(link):
                            modules.append({'text': text, 'link': link})

        title_tag = soup.find('title').get_text(strip=True)
        product_name = title_tag.split(' - ')[0]
        product_img = soup.find('meta', property='og:image')['content']

        products.append({
            'name': product_name,
            'url': product_url,
            'img': product_img,
            'modules': modules
        })
```

```python
products
```

```python
import json

with open('products.json', 'w') as file:
    json.dump(products, file, indent=4, ensure_ascii=False)
```
